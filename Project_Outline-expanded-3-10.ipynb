### Read in Haplorhini range data 

We want to know the earliest and latest appearance of the fossil. Find first and last occurrence of every species! For each species get uniques and stick them into a dictionary with the earliest and latest occurence. We want to end up with a file  that is csv formatted that includes only the occurrence records (no headers). use shell command.

````python
## open the file

infile = "/home/eeb177-student/Desktop/eeb-177/eeb-174-final-project/formattedapelist.csv"
with open(infile, 'r') as ff:
    homin_recs = ff.readlines()
````

````python
# use default dict
from collections import defaultdict
species_ranges = defaultdict(list)


# make a dict
for line in homin_recs:
    species = line.split(",")[5]
    minage = line.split('","')[14]
    maxage = line.split('","')[15]
    mean_age = (float(minage) + float(maxage))/2
    #print(mean_age)
    
    species_ranges[species].append(mean_age) # add speices as keyes and mean ages as elements tot he list of ages
````

````python
output = open("homin-ranges.csv", "w")

#get oldest and youngest
#species_ranges.values()
for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    species = key
    outline= "{},{},{},{}\n".format(genus, species, minage, maxage)
    #output.write(outline)
    #print(genus, key, minage, maxage)
    #output.write(genus + "," + key + ", ")
    print(outline)
````

````python
in_file = open("/home/eeb177-student/Desktop/eeb-177/eeb-174-final-project/haplomeasurewidth.csv")
# read in the first line
# The first line is a header that we don't want in dictionaries
in_file.readline()
# set up empty dictionaries
sizes_width = {}
sizes_mass = {}
sizes_length = {}

# iterate through every line in the file
for line in in_file:
    #print(line)
    
    # It's a tab delimited file, so we can split the contents up into 
    # the component rows
    items = line.split(",")
    
    # Save some important pieces of information as variables
    specimen_id = items[0]
    width_avg = float(items[1])
    
    
    # Use the variables to populate our three empty lists
    sizes_width[specimen_id] = width_avg
    
````

````python

import matplotlib.pyplot as plt


# Most plotting happens on lists, not dictionaries-
# so, let's convert our 1994 DBH meaasurements into a list
width_avg = list(sizes_width.values())

# We're ready to make a histogram!
plt.hist(width_avg)
plt.xlabel('Width (inch)')
# a y-axis label
plt.ylabel('Frequency')
# A main title for the graph
plt.title("Width Size distribution of Haplorhini Specimen")
plt.show()
````

````python
# Add a log-axis to our histogram
plt.hist(width_avg, log = True)
# an X-axis label
plt.xlabel('Width (inch)')
# a y-axis label
plt.ylabel('Frequency')
# A main title for the graph
plt.title("Width Size distribution of Haplorhini Specimen")
# Let's look!
plt.show()
````

````python
f, (ax1, ax2) = plt.subplots(1, 2,sharex = True)
ax1.hist(width_avg, log = False)
ax1.set_title('Unlogged')
ax2.hist(width_avg, log = True)
ax2.set_title('Logged')
plt.show()
````
